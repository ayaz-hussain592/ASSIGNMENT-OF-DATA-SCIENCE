```python
import pandas as pd
import numpy as np
data = pd.read_csv("../netflix_titles.csv")
data1 = pd.read_csv("../netflix_data.csv")
```

# :::::::: TASK-1 :::::
ADD NEW COLUMN AS "Number of Cast"

```python
data.isnull().sum()
```

```python
# *******REPLACING "NaN" values of Cast Column by zero*******

data.replace(to_replace = np.nan , value = 0 , inplace = True)


# ******ADDING NEW COLUMN IN THE GIVEN DATA********
def set_new_colm(value):
    if value == 0:
        return 0
    else:
        array = value.split(",")
        length = len(array)
        return length
    
data["Num_of_Cast"] = data["cast"].apply(set_new_colm) 
```
